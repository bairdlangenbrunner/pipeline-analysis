```python
import pandas
import geopandas
import matplotlib.pyplot as mp
import numpy

import cartopy
import shapely
```

```python
pipelines_df = pandas.read_csv('../Pipelines_Current-Pipelines.csv')
```

```python
pipelines_df.head()
```

```python
pipelines_mls = pipelines_df['WKT Format']
```

```python
type(pipelines_mls)
```

```python
pipelines_mls[0]
```

```python
mls = shapely.wkt.loads(pipelines_mls[0])
```

```python
geometry_list = []

for idx,entry in enumerate(pipelines_mls):
    if str(entry)=='#REF!': pass
        #geometry_list.append(numpy.nan)
    elif str(entry)=='MULTILINESTRING ((NA Bidirectionalityupgradeonly))': pass
        #geometry_list.append(numpy.nan)
    elif str(entry)=='nan': pass
        #geometry_list.append(numpy.nan)
    else:
        geometry_list.append(shapely.wkt.loads(entry))
```

```python
type(mls)
```

```python
proj = cartopy.crs.PlateCarree()

fig = mp.figure(figsize=(15,8))

ax = fig.add_subplot(111, projection=proj)

ax.coastlines()

ax.add_geometries([mls], crs=proj, facecolor='None', edgecolor='darkred', linewidth=2)
```

```python
proj = cartopy.crs.PlateCarree()

fig = mp.figure(figsize=(15,8))

ax = fig.add_subplot(111, projection=proj)

ax.coastlines()

ax.add_geometries(geometry_list, crs=proj, facecolor='None', edgecolor='darkred', linewidth=2)
```
